```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler

# Reading the CSV file into a DataFrame
df = pd.read_csv('creditcard.csv')

# Displaying the first five rows of the DataFrame
print(df.head())

# Getting descriptive statistics
print(df.describe())

# Getting information about the DataFrame
print(df.info())

# Example of flattening an array of axes (if you have multiple subplots)
fig, ax = plt.subplots(2, 2)  # Example: creating a 2x2 grid of subplots
ax = ax.flatten()  # Flattening the array of axes objects

# Display all columns of the DataFrame
pd.options.display.max_columns = None

# Display all rows of the DataFrame
pd.options.display.max_rows = None

# 'shape' function gives the total number of rows and columns in the data
print(df.shape)

# Display the data types of each column
print(df.dtypes)

# Displaying index information
print("\nIndex information:")
print(f"Index type: {type(df.index)}")
print(f"Index range: {df.index}")
print(f"Index values: {df.index.values}")

# Displaying column names
print("\nColumn names:")
print(df.columns)

# Basic dataset information
print(df.head())
print(df.info())
print(df.describe())

# Plot histograms for numeric variables in df
df.hist(figsize=(14, 10))
plt.tight_layout()
plt.show()

# Step 5: Plot the countplot of the variable 'price_range'
if 'price_range' in df.columns:
    sns.countplot(x=df['price_range'])

    # Calculate class frequencies from df
    class_frequency = df['price_range'].value_counts()

    # Add text annotations for percentages
    total = len(df['price_range'])
    plt.text(x=-0.05, y=class_frequency[0], 
             s=str(round((class_frequency[0]) * 100 / total, 2)) + '%')
    plt.text(x=0.95, y=class_frequency[1], 
             s=str(round((class_frequency[1]) * 100 / total, 2)) + '%')
    plt.text(x=1.95, y=class_frequency[2], 
             s=str(round((class_frequency[2]) * 100 / total, 2)) + '%')
    plt.text(x=2.95, y=class_frequency[3], 
             s=str(round((class_frequency[3]) * 100 / total, 2)) + '%')

    plt.title('Count Plot for Target Variable (price_range)', fontsize=15)
    plt.xlabel('Target Variable', fontsize=15)
    plt.ylabel('Count', fontsize=15)
    plt.show()
else:
    print("The column 'price_range' does not exist in the dataset.")

# Visualize the class distribution (fraud vs non-fraud)
if 'Class' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Class', data=df)
    plt.title('Class Distribution (0: Non-Fraud, 1: Fraud)')
    plt.show()
else:
    print("The column 'Class' does not exist in the dataset.")

# Visualize the distribution of 'Amount'
if 'Amount' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df['Amount'], bins=50, kde=True)
    plt.title('Distribution of Transaction Amounts')
    plt.show()
else:
    print("The column 'Amount' does not exist in the dataset.")

# Visualize Transaction Time vs Amount
if 'Time' in df.columns and 'Amount' in df.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Time'], df['Amount'], alpha=0.5)
    plt.title('Transaction Time vs Amount')
    plt.xlabel('Time')
    plt.ylabel('Amount')
    plt.show()
else:
    print("The columns 'Time' and/or 'Amount' do not exist in the dataset.")

# Correlation Heatmap (Scaling 'Amount' and 'Time' for better visualization)
if 'Amount' in df.columns and 'Time' in df.columns:
    scaler = StandardScaler()
    df_scaled = df.copy()
    df_scaled[['Amount', 'Time']] = scaler.fit_transform(df[['Amount', 'Time']])

    # Plot heatmap of correlations
    plt.figure(figsize=(12, 8))
    sns.heatmap(df_scaled.corr(), cmap='coolwarm', annot=False, linewidths=0.5)
    plt.title('Correlation Heatmap of Features')
    plt.show()
else:
    print("The columns 'Amount' and/or 'Time' do not exist in the dataset.")

# Boxplot of 'Amount' by Class
if 'Class' in df.columns and 'Amount' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Class', y='Amount', data=df)
    plt.title('Boxplot of Amount by Class (0: Non-Fraud, 1: Fraud)')
    plt.show()
else:
    print("The columns 'Class' and/or 'Amount' do not exist in the dataset.")

# Distribution of the 'V' features (V1 to V28)
v_features = [col for col in df.columns if col.startswith('V')]

if len(v_features) > 0:
    plt.figure(figsize=(14, 20))
    for i, col in enumerate(v_features):
        plt.subplot(7, 4, i + 1)
        sns.histplot(df[col], bins=30, kde=False)
        plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.show()
else:
    print("No columns starting with 'V' found in the dataset.")

# Visualization 1: Pairplot for selected features
selected_columns = ['V1', 'V2', 'V3', 'V4', 'Amount', 'Class']
sns.pairplot(df[selected_columns], hue='Class', palette='coolwarm')
plt.title('Pairplot of Selected Features')
plt.show()

# Visualization 2: Correlation matrix with annotations
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Matrix with Annotations')
plt.show()

# Visualization 3: Class distribution pie chart
if 'Class' in df.columns:
    class_counts = df['Class'].value_counts()
    labels = ['Non-Fraud', 'Fraud']
    plt.pie(class_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#3498db', '#e74c3c'])
    plt.title('Class Distribution (Fraud vs Non-Fraud)')
    plt.show()

# Visualization 4: Scatter plot with hue for fraud
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Time', y='Amount', hue='Class', data=df, palette='coolwarm', alpha=0.7)
plt.title('Transaction Time vs Amount (Hue by Fraud Class)')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()

# Visualization 5: Violin plot for distribution of Amount by Class
plt.figure(figsize=(8, 6))
sns.violinplot(x='Class', y='Amount', data=df, palette='coolwarm', split=True)
plt.title('Violin Plot of Amount by Class (0: Non-Fraud, 1: Fraud)')
plt.show()

# Visualization 6: Distribution of V1 feature by Class
plt.figure(figsize=(8, 6))
sns.kdeplot(df[df['Class'] == 0]['V1'], label='Non-Fraud', fill=True)
sns.kdeplot(df[df['Class'] == 1]['V1'], label='Fraud', fill=True)
plt.title('Distribution of V1 by Class')
plt.xlabel('V1')
plt.legend()
plt.show()

# Visualization 7: Heatmap of missing values
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values')
plt.show()

# Visualization 8: Boxplot for V features
plt.figure(figsize=(16, 6))
sns.boxplot(data=df[v_features], palette='Set2')
plt.xticks(rotation=90)
plt.title('Boxplot for V Features')
plt.show()
```
